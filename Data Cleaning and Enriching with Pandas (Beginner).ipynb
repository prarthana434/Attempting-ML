```python
df1 = df1.drop(["WidgetName", "IP Address", "ReferralURL Parameters", "Landing Page URL Parameters"], axis = 1)

int_nulls = 0

df = df1.fillna( {"ReferralURL" : "Direct Source", 
            "ReferralURL Parameters" : "Does not apply",
            "SearchQuery" : "Direct Source",
            "Keyword" : "Direct Source",
            "CallStatus" : "Not yet contacted",
            "AddressScore" : int_nulls,
            "PhoneScore" : int_nulls
           })

df = df.astype({
    "FirstName" : "category",
    "Email" : "category",
    "VendorLeadID" : "category",
    "CallStatus" : "category",
    "PublisherZoneName" : "category",
    "PublisherCampaignName" : "category",
    "AdvertiserCampaignName" : "category",
    "State" : "category",
    "DebtLevel" : "category",
    "Partner" : "category",
    "ReferralDomain" : "category",
    "MarketingCampaign" : "category",
    "AdGroup" : "category",
    "Keyword" : "category",
    "SearchQuery" : "category",
    "ReferralURL" : "category",
    "LandingPageURL" : "category",
    "AddressScore" : "int",
    "PhoneScore" : "int"
})
```

```python
df["Lead Created Time"] = df["LeadCreated"].dt.time
df["Lead Created Date"] =  df["LeadCreated"].dt.strftime("%m-%d-%Y")
df["Lead Month"] = df["LeadCreated"].dt.month
df["Lead Year"] = df["LeadCreated"].dt.year
df["Day"] = df["LeadCreated"].dt.day
df["Weekday"] = df["LeadCreated"].dt.weekday
df["Lead Week"] = df["LeadCreated"].dt.isocalendar().week
df["Month Name"] = df["LeadCreated"].dt.month_name()
df["Lead Quarter"] = df["LeadCreated"].dt.quarter

def dayname(a):
    if a["Weekday"] == 1:
        return "Sunday"
    elif a["Weekday"] == 2:
        return "Monday"
    elif a["Weekday"] == 3:
        return "Tuesday"
    elif a["Weekday"] == 4:
        return "Wednesday"
    elif a["Weekday"] == 5:
        return "Thrusday"
    elif a["Weekday"] == 6:
        return "Friday"
    else:
        return "Saturday"

df["Day Name"] = df.apply(dayname, axis = 1)
```

```python
df.drop_duplicates(subset = {"LeadCreated", "VendorLeadID"})
```

```python
df.groupby(["Lead Month", "Month Name", "Lead Week"]).agg(leads = ("VendorLeadID", "nunique")).reset_index().sort_values(by = ["Lead Week", "Lead Month"], ascending = [True, True])
```
